```python
max_y_value_for_all_data = []

for index, isotherm in enumerate(X):
    y_max = isotherm[-1][1]
    bet_value = y[index]
    data_name = info[index][0]
    sample_name = info[index][1]
    graph_name = info[index][2]
    max_y_value_for_all_data.append(y_max)  
    if(bet_value > 1500 and y_max > 1250):
        print(f"no: {index}   {data_name} ->  BET: {bet_value}  |  y_max={y_max}         sample_name: {sample_name}  graph: {graph_name}")
plt.scatter(max_y_value_for_all_data, y, color='blue', label='Dane')
plt.title('Zaleznosc powierzchni wlasciwej od ilosci zaabsorbowanego gazu')
plt.xlabel('y max value [cm3/g]')
plt.ylabel('BET value [m2/g]')
# plt.legend()
plt.show()
```
